## 3. Contexto del Problema

### Descripci√≥n del Problema:

La accidentalidad vial es una de las principales causas de muerte y lesiones en Colombia. Seg√∫n la Agencia Nacional de Seguridad Vial (ANSV), en 2024 se registraron 8,271 muertes por accidentes de tr√°nsito en el pa√≠s, siendo los motociclistas el grupo m√°s afectado con el 61.61% de las fatalidades.

En **Bucaramanga**, entre 2012 y 2023 se registraron **39,193 accidentes de tr√°nsito**, con la siguiente distribuci√≥n:
- **50%** solo da√±os materiales
- **48.4%** con personas heridas  
- **1.6%** con personas fallecidas (609 accidentes mortales)

**Hallazgos Preliminares:**
- El **70%** de los accidentes ocurren durante el d√≠a
- Los **autom√≥viles** (30,620) y las **motos** (24,559) son los veh√≠culos m√°s involucrados
- El barrio **Centro** concentra el 9.85% de todos los accidentes
- Se observ√≥ una ca√≠da significativa en 2020 (COVID-19): de ~3,900/a√±o a 2,261
- Hay una inconsistencia en los datos: "Con muertos" vs "Con Muertos" (a corregir en limpieza)

**¬øQu√© se busca analizar?**
- Patrones espaciales y temporales de la accidentalidad en Bucaramanga
- Factores asociados a la gravedad de los accidentes
- Tipos de veh√≠culos m√°s involucrados en accidentes graves
- Identificaci√≥n de barrios y comunas de alto riesgo
- Relaci√≥n entre variables: hora, d√≠a, tipo de veh√≠culo, gravedad
- Evoluci√≥n temporal de la accidentalidad (2012-2023)

**¬øQu√© se busca resolver?**
- Reducir la tasa de accidentalidad mediante intervenciones basadas en evidencia
- Optimizar la asignaci√≥n de recursos de prevenci√≥n y control en zonas cr√≠ticas
- Mejorar la infraestructura vial en puntos de alta accidentalidad
- Dise√±ar campa√±as de educaci√≥n vial focalizadas (por ejemplo: seguridad en motos)
